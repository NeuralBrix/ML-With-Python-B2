```python
# mathematical set operations

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
```

```python
# Union
set_union = set_a.union(set_b)
print("Union:", set_union)
print("using pipe char", set_a | set_b) 
```

```python
# intersection
set_intersection = set_a.intersection(set_b)
print("Intersection:", set_intersection)
print("using ampersand char", set_a & set_b)
```

```python
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
```

```python
#Difference
set_difference = set_b.difference(set_a)  
print("Difference (A-B):", set_difference)
print("using minus char", set_b - set_a)
```

```python
#symmetric difference
set_sym_diff = set_a.symmetric_difference(set_b)
print("Symmetric Difference:", set_sym_diff)
print("using caret char", set_a ^ set_b) #caret(^)
```
