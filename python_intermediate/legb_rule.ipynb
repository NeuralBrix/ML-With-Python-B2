```python
x = "global"
def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print("Printed :", x)
    print("Printed :", x)
    inner()
outer() 
print("Printed :", x)
```

```python
x = 5
def change():
    global x
    x = 10
    print("Inside change():", x)
change()
print("Outside change():", x)
```
