```python
#positional only parameter
def area(radius,/,unit):
    pi = 3.1416
    result = str(pi * radius * radius) + " " + unit
    return result
```

```python
#positional argument
area(10,"sq cm") 
```

```python
# positional or keyword argument
def volume(radius, height, unit):
    pi = 3.1416
    result = str(pi * radius * radius * height) + " " + unit
    return result
```

```python
volume(10, 5, unit = "cubic cm") #positional argument
```

```python
volume( 5, height=10, unit="cubic cm") #keyword argument
```

```python
volume(10, 5, unit="cubic cm") #mixed argument # valid                                             
```

```python
volume(10, radius=5, "cubic cm")
```

```python
# var positional argument
def sum_all(*args, sample, sample2):
    total = 0
    for num in args:
        total += num
    print("Sample value is:", sample)
    return total    
```

```python
sum_all(1, 2, 3,3,5,4,4,5,34,34,34,3,4, sample="mehedi", sample2="hasan")  
```
